```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# grfico 1 - Linha
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [13, 50, 348, 296, 573, 196, 760, 496, 983, 486, 951, 30]

axs[0,0].plot(meses, vendas)
axs[0,0].set_xlabel('Ms')
axs[0,0].set_ylabel('Vendas em Milhares de Reais')
axs[0,0].set_title('Crescimeto das Vendas ao Longo do Ano')

# grfico 2 - Barras
cidades = ['So Paulo', 'Rio de Janeiro', 'Salvador']
temperaturas = [20, 30, 40]

axs[0,1].bar(cidades, temperaturas)
axs[0,1].set_xlabel('cidades')
axs[0,1].set_ylabel('Temperatura Mdia em Grus Clcios')
axs[0,1].set_title('Comparando Temperaturas Mdias')

# grfico 3 - Pizza
categorias = ['Eletrnicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 39, 15, 10]
destaques = [0, 0, 0, 0.1]

axs[1,0].pie(quantidades, labels=categorias, autopct='%.1f')
axs[1,0].set_title('Distribuio percentual de categorias de produtos')

# grfico 4 - Barras Empilhadas
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

axs[1,1].bar(anos, vendas_online, label='online', )
axs[1,1].bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
axs[1,1].set_xlabel('Ano')
axs[1,1].set_ylabel('Vendas(Em milhes de reais)')
axs[1,1].set_title('Vendas online vs Lojas')
axs[1,1].set_xticks(anos)
axs[1,1].legend()

# ajustes finais
fig.suptitle('Evoluo de Vendas', fontsize=16)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses = ['jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

marcadores = np.where(vendas == np.min(vendas), 'x', '')

# Palheta de Cores
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

fig, ax = plt.subplots(figsize=(8, 4))

# Grfico base
ax.plot(meses, vendas)
                      
ax.axvspan(6, 11, color=cores[1], alpha=0.15)                      

# Marcadores e Anotaes
for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor='red'
        tamanho = 80
    else:
        cor = 'blue'
        tamanho = 30
    ax.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)
    ax.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5, 5), textcoords='offset points')
    
#ax.annotate('Menor valor', color='red', xy=(2, 78), xytext=(2.2, 78))
# Destaque de Regio
ax.axvspan(6, 11, alpha=0.2, color=cores[1])
ax.text(6.5, 80, 'Recuperao de Crescimento', color='green', fontsize=9, weight='bold')
ax.set_frame_on(False)
ax.tick_params(axis='x', length=0)
ax.set_yticks([])

ax.set_xlabel('Ms')
ax.set_ylabel('Receita em Milhares')
ax.set_title('Crescimento das Vendas ao Longo do Ano')


plt.show()
```

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
dados_ordenados = sorted(zip(quantidades, produtos))
quantidades_ordenadas = [x[0] for x in dados_ordenados]
produtos_ordenados = [x[1] for x in dados_ordenados]

import matplotlib.pyplot as plt
from cycler import cycler

cores = plt.get_cmap('tab20').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

fig, ax = plt.subplots()

#rotulos = [valor if valor == max(quantidades_ordenadas) else '' for valor in quantidades_ordenadas]
barras = ax.barh(produtos_ordenados, quantidades_ordenadas)

ax.bar_label(barras, labels=[max(quantidades_ordenadas) if x == max(quantidades_ordenadas) else '' for x in quantidades_ordenadas], padding=3, fontsize=10, fontweight='bold')

ax.set_frame_on(False)

ax.set_xticks([])

ax.tick_params(axis='y', length=0)

for i, barra in enumerate(barras):
    if i != quantidades_ordenadas.index(max(quantidades_ordenadas)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.2)

ax.set_title('Quantidades de Produtos Vendidos')

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

fig, ax = plt.subplots()

barras_online = ax.bar(anos, vendas_online, label='online')
barras_loja = ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
ax.bar_label(barras_online, labels=vendas_online, label_type='center', fontweight='bold')
ax.bar_label(barras_loja, labels=vendas_loja, label_type='center',fontweight='bold')

#forar os anos a serem exibidos como inteiros
ax.set_xticks(ticks=anos)

ax.set_frame_on(False)
ax.tick_params(axis='x', length=0)
ax.set_yticks([])

ax.set_ylabel('Vendas(Em milhes de reais)')
ax.set_title('Vendas online vs Lojas')
ax.legend()


plt.show()
```
